# Modelling

All of science relies on models. 

A model allows us to reason logically from assumptions about a system‚Äôs structure to conclusions about what we should see, would see under certain conditions, or will see if we intervene.

We turn to models when the part of the world we‚Äôre interested in is difficult to work with directly. For example, the system might be:
- too large or too small
- too expensive to test
- too hard to manipulate
- or simply too messy to make sense of all at once

Much of the art of the scientist lies in formulating, manipulating, and revising models ‚Äî all with the understanding that the model does *not*, and *should not*, attempt to capture the full richness of the world. We don't want an exact replica of the system - we already have that, which is why we need a model! (Shameless plug: [My TedX talk](https://www.youtube.com/watch?v=Pgm65hYglpg))

Modelling in general (and modelling of complex systems in particular) requires:
- a growing comfort with approximation
- the use of partial descriptions
- an acceptance of generalisations with limited scope

These concessions aren't weaknesses ‚Äî they‚Äôre what make modelling possible, and more importantly, they are what make modelling useful!

### Planetary orbits

**Kepler's observation:** The orbit of every planet is an ellipse with the Sun at one of the two foci. (This, and his other two laws, were found by analysing astronomical data.)

**The model:** Isaac Newton proved Kepler was right by following a sort of 'recipe'... 
   1. he appealed to a general principle - the *law* of gravitation, which can be 
   2. written as a *differential equation* and 
   3. *analytically solved* this to give an elliptic solution
   4. *QED*. 

<div style="background-color: #ffff88; padding: 1em; border-left: 5px solid #fff102;">
    Mathematical detail of this process for on board? Depending on level of students present or set as optional homework exercise?<br><br>

1. **Newton's Law of Universal Gravitation**
   Newton postulated that every mass attracts every other mass with a force given by:

   $$
   \vec{F} = -\frac{GMm}{r^2} \hat{r}
   $$

   where $G$ is the gravitational constant, $M$ is the mass of the Sun, $m$ the mass of the planet, $r$ the distance between them, and $\hat{r}$ a unit vector pointing from the planet to the Sun.

2. **Second Law of Motion ‚Üí Differential Equation**
   Newton‚Äôs Second Law $\vec{F} = m\vec{a}$ gives:

   $$
   m\vec{a} = -\frac{GMm}{r^2} \hat{r}
   \quad \Rightarrow \quad
   \vec{a} = \ddot{\vec{r}} = -\frac{GM}{r^2} \hat{r}
   $$

   This is a second-order vector differential equation for the planet's position $\vec{r}(t)$.

3. **Solution in Polar Coordinates**
   Exploiting conservation of angular momentum and using polar coordinates $(r, \theta)$, the equation becomes:

   $$
   \frac{d^2u}{d\theta^2} + u = \frac{GM}{h^2}
   \quad \text{where } u = \frac{1}{r}, \quad h = r^2 \frac{d\theta}{dt}
   $$

   This is a linear second-order ODE.

4. **Elliptical Orbit Solution**
   The general solution is:

   $$
   u(\theta) = \frac{GM}{h^2} \left(1 + e \cos(\theta) \right)
   \quad \Rightarrow \quad
   r(\theta) = \frac{h^2/GM}{1 + e \cos(\theta)}
   $$

This is the equation of a conic section ‚Äî an **ellipse** when $0 < e < 1$, confirming Kepler‚Äôs empirically observed result (his first law).

</div>

This mathematical derivation gives rigor and a sense of absolute and unwavering proof of the observed phenomena, so people tended to like this.

The study of complex systems traces back to four foundational scientific developments of the 19th century. These historical prequels laid the groundwork for our modern understanding of *systems composed of many interacting parts*, particularly where the behaviour of the whole cannot be easily deduced from the parts.

1. **Entropy and Thermodynamics**  
   *Key contributors:* Carnot, Clausius, Maxwell, Boltzmann, Gibbs  
   From the second law of thermodynamics to statistical mechanics, these pioneers helped us understand the delicate interplay between **order and disorder**, and the constraints they impose on the production of work in physical systems.

2. **Evolution and Adaptation**  
   *Key contributors:* Darwin, Wallace, Mendel  
   With natural selection and genetic inheritance, biology revealed itself to be a system of **adaptive feedback**, driven by variation, competition, and replication. The mathematics of population dynamics and evolutionary game theory build on this legacy.

3. **Nonlinear Dynamics**  
   *Key contributors:* Maxwell, Poincar√©  
   The study of *nonlinear differential equations*, stability, and chaos emerged as tools for understanding how deterministic systems can produce unpredictable behaviour ‚Äî from weather to cardiac rhythms.

4. **Computation and Logic**  
   *Key contributors:* Babbage, Lovelace, Boole  
   The groundwork for algorithmic thinking, symbolic logic, and digital computation was laid. These ideas are now central to understanding the information processing and decision-making capabilities of complex systems.

These pillars still underpin the modern mathematical language of complexity. In addition to them, the 20th century brought conceptual advances from:
- **Cybernetics** (feedback, control)
- **General Systems Theory**
- **Game Theory** (strategic interaction)

As a relatively new field the boundaries and key terms are hotly debated. Personally, I like Hiroki Sayama's topical clusters. These are often overlapping perspectives that reflect the core themes, tools, and phenomena in Complex Systems:

Classical modelling is highly effective at describing systems at two opposite extremes:

- **Simple systems**  
  These involve a small number of interacting components governed by well-understood rules. Their behaviour is highly predictable and can be described using exact mathematical equations.
  *Example:* The motion of planets is accurately modelled by Newton‚Äôs laws of motion and gravitation.

- **Complicated (but disorganised) systems**  
  These consist of many components, but interactions are weak, random, or statistically uniform. Their complexity arises from scale, not structure, and they are amenable to averaging and aggregation.
  *Example:* A gas made up of billions of atoms can be described by statistical mechanics, which ignores individual behaviour in favour of macro-level regularities.

In both cases, classical systems are typically:
**closed**, **reversible**, **symmetry-dominated**, and **predictable**.
They lend themselves to reductionist approaches, where behaviour can be derived from first principles.

At one extreme, we find systems whose behaviour is determined by fundamental symmetries and universal laws, requiring only a few initial conditions. At the other, the system is so complicated that individual details become irrelevant ‚Äî patterns emerge only at the aggregate level, and these can be captured through statistical or probabilistic descriptions.

But classical modelling can struggle when attempting to disentangle complex systems sitting in the awkward space between these extremes... and that's exactly where most of the interesting stuff lies!

## Complex systems (the systems)
The actual systems themselves ‚Äî made up of many interacting parts that influence one another in non-trivial ways.

These interactions can produce **emergent behaviour** ‚Äî patterns, structures, or dynamics not present in any individual component.

For example, bird flocks exhibit coordinated motion that seems almost choreographed ‚Äî turning, twisting, and dispersing as one. But there is no leader and no global control mechanism. Each bird follows a few simple local rules: maintain distance, align with neighbours, and avoid collisions. The flock‚Äôs elegant motion emerges from these local interactions, i.e. it is a property of the whole, not found in any single bird or rule.

Emergence is one of the key features of a complex system. There are many others listed below. A complex system is likely not exhibit all of these features, but many will share a recognisable subset of traits that serve as signposts that we're dealing with a system whose behaviour cannot be understood by studying parts in isolation.

##### Key features of complex systems

- **Interactions:** across spatial and/or temporal scales. 
- **A large number of components** (but not too large)
- **Nonlinearity:** Relationships between inputs and outputs, or between components within the system, are complex and often not linear.
- **Heterogeneity:** Diversity across the components in the system such as different attributes, behaviours, or states.
- **Self-organising:** with rules that amplify fluctuations into coordinate space-time patterns
- **Emergence:** A theory of the system at some scale or level of organisation is qualitatively different from the lower level scale. i.e *the whole is more than the sum of its parts* - Aristole. Society is more than a collection of individuals, consciousness is more than a collection of neurons, traffic is more than a collection of cars. This is a behavioural phenomena at the system level
- **Multiscale dynamics:** Dynamics occur across multiple scales. The system often lacks of a characteristic scale. How do we go from the dynamics of neurons to brain function and behaviour?
- **Far from equilibrium:** complex systems usually exist in an open environment from which individual components can both consume and dissipate energy, allowing for time reversal symmetry to break. Some complex systems are referred to as 'active matter', if the individuals in the matter are moving somehow then they may be referred to as 'self-propelled particles'. Individuals may also have an ability to express a strategy, adapting, coordinating, optimising as they make decisions in an evolving environment.
    - As a related but somewhat rabbit-holey tangential: [Does free-will violate physics?](https://www.youtube.com/watch?v=rohgVwQ57uM&t=393s)
- **Strong transitions:** Abrupt shifts can occur as systems pass critical points (called bifurcation or tipping points). This may continue to produce a cascade, where one failure (or success) can lead to cascading failures (or successes) that may have catastrophic (or beneficial) consequences.
- **Memory:** The processes that make up the world are dynamic: the past informs on the future and individual moments cannot be assumed to be decoupled from the history that lead up to that point. Memory in a complex system isn‚Äôt stored in a single location ‚Äî it is distributed across the system. As a result, the system has a history, and that history plays a crucial role in shaping its behaviour.
- **Feedback Loops:** Both negative (dampening), and positive (amplifying) feedback are almost always found in complex systems. Negative feedback tends to move a system towards equilibria that are stable, predictable, and resistant to change, whilst positive feedback moves a system away from equilibria and promotes further disruption. Whether a feedback is classified as positive or negative is based on its influence on the patterns and dynamics.
- **Dissipative:** dynamics that are irreversible and break time-reversal symmetry
- **Adaptive:** capture of information by a system from the environment that promotes persistence and increases multiplicity
- More traits:
    - path dependence
    - power laws
    - universality
    - rich phase diagrams
    - produce and destroy diversity
    - inherently prone to collapse
    - resilient
    - stochastic
    - multilayer networks
    - evolution and co-evolution
    - ...

### Dynamics
- **Neighbourhood:** <font color='red'>Moore, not including self</font>. Entirely summarised by similarity ratio $\frac{n_s}{n}$, where $n_s$ are the number of similar neighbours in the neighbourhood and $n = \{1,\ldots,8\}$ is the number of total neighbours
- **Dynamics:** Based on local neighbourhood similarity ratio and a threshold $\theta_r=50\%$. Agent is unhappy and will want to move if <font color='red'> $\frac{n_s}{n}<\theta_r$</font>. Randomly selected unhappy agent will move to random empty location

````python
def find_dissatisfied_indices(grid, threshold_percentage=50):
    """Returns (i, j) coordinates of agents dissatisfied with their neighbourhood."""
    dissatisfied = []
    rows, cols = grid.shape

    for i in range(rows):
        for j in range(cols):
            if grid[i, j] == 0:
                continue  # Skip empty cells

            # Extract 3x3 neighbourhood with bounds checking
            neighbourhood = grid[max(0, i - 1):min(rows, i + 2),
                                 max(0, j - 1):min(cols, j + 2)]

            # Count similar and empty neighbours
            total_neighbours = np.count_nonzero(neighbourhood) - 1  # exclude self
            similar_neighbours = np.count_nonzero(neighbourhood == grid[i, j]) - 1

            if total_neighbours == 0:
                local_mix = 0
            else:
                local_mix = 100 * similar_neighbours / total_neighbours

            if local_mix < threshold_percentage:
                dissatisfied.append((i, j))

    return dissatisfied
````

````python
def plot_grid_with_dissatisfied(grid, threshold_percentage=50, title=None, ax=None, show_legend=True, find_dissatisfied_indices=None):
    """
    Plot the Schelling grid with dissatisfied agents marked.

    Parameters:
    - grid: 2D numpy array
    - threshold_percentage: dissatisfaction threshold
    - title: optional title string
    - ax: matplotlib axis (optional)
    - show_legend: whether to display legend
    - find_dissatisfied_indices: function to identify dissatisfied agents (must return (i, j) tuples)
    """
    assert find_dissatisfied_indices is not None, "Must provide find_dissatisfied_indices function"

    cmap = ListedColormap(["white", "#FFD700", "#4682B4"])

    if ax is None:
        fig, ax = plt.subplots(figsize=(4, 4))

    ax.imshow(grid, cmap=cmap, vmin=0, vmax=2)
    ax.set_xticks([]); ax.set_yticks([])
    ax.grid(which="both", color="lightgrey", lw=0.8, alpha=0.5)

    if title:
        ax.set_title(title, pad=12)

    # Overlay emoji on dissatisfied agents
    dissatisfied = find_dissatisfied_indices(grid, threshold_percentage)
    for i, j in dissatisfied:
        ax.text(j, i, 'üòû', ha='center', va='center', fontsize=100 / grid.shape[0], color='black')

    if show_legend and ax is not None:
        patch_empty = mpatches.Patch(color="white", label="Empty")
        patch_group1 = mpatches.Patch(color="#FFD700", label="Group 1")
        patch_group2 = mpatches.Patch(color="#4682B4", label="Group 2")
        patch_dissatisfied = mpatches.Patch(color='none', label=f"üòû Dissatisfied (< {threshold_percentage}%)")
        ax.legend(
            handles=[patch_empty, patch_group1, patch_group2, patch_dissatisfied],
            loc="center left", bbox_to_anchor=(1.02, 0.5),
            frameon=False, fontsize=12
        )

    if ax is None:
        plt.tight_layout()
        plt.show()
````

````python
def run_one_simulation(grid, find_dissatisfied_indices=None, get_mean_similarity=None, threshold_percentage=50,
                       force_iterate=False, collect_data=False, collect_frames=False, sim_rng=None, max_iterations=200):
    """Simulates Schelling model.

    Options:
    - collect_data: returns MSR and dissatisfaction over time
    - collect_frames: returns frames and moves for animation

    Returns:
    - If both False: final grid only
    - If collect_data: (msr, dissatisfaction_counts, grid)
    - If collect_frames: (dissatisfaction_counts, grid, frames, moves)
    - If both: (msr, dissatisfaction_counts, grid, frames, moves)
    """
    assert find_dissatisfied_indices is not None, "Must provide a dissatisfaction function"

    # Ensure get_mean_similarity is passed if needed
    if collect_data and get_mean_similarity is None:
        raise ValueError("Must provide get_mean_similarity when collect_data=True")

    # Otherwise provide dummy if not needed
    if get_mean_similarity is None:
        def get_mean_similarity(grid): return None
            
    if sim_rng is None:
        sim_rng = default_rng()
        print("WARNING: new random seed generated")
            
    grid = grid.copy()
    iterations = 0
    prev_grid = grid.copy()

    msr = [get_mean_similarity(grid)] if collect_data else None
    dissatisfaction_counts = [len(find_dissatisfied_indices(grid, threshold_percentage))] if (collect_data or collect_frames) else None
    frames = [grid.copy()] if collect_frames else None
    moves = [None] if collect_frames else None

    while iterations < max_iterations:
        dissatisfied = find_dissatisfied_indices(grid, threshold_percentage)

        if not dissatisfied:
            if not force_iterate:
                break
            # No move ‚Äî just continue collecting static data
            iterations += 1
            if collect_data:
                msr.append(get_mean_similarity(grid))
            if dissatisfaction_counts is not None:
                dissatisfaction_counts.append(0)
            if collect_frames:
                frames.append(grid.copy())
                moves.append(None)
            continue

        # Pick a dissatisfied agent (i, j) directly
        i, j = sim_rng.choice(dissatisfied)

        empties = np.argwhere(grid == 0)
        if empties.size == 0:
            break

        m, n = empties[sim_rng.integers(len(empties))]
        grid[m, n], grid[i, j] = grid[i, j], 0

        iterations += 1

        if collect_data:
            msr.append(get_mean_similarity(grid))
        if dissatisfaction_counts is not None:
            dissatisfaction_counts.append(len(find_dissatisfied_indices(grid, threshold_percentage)))
        if collect_frames:
            frames.append(grid.copy())
            moves.append(((i, j), (m, n)))

        if not force_iterate and np.array_equal(prev_grid, grid):
            break
        prev_grid = grid.copy()

    # Return based on collection options
    if collect_data and collect_frames:
        return msr, dissatisfaction_counts, grid, frames, moves
    elif collect_data:
        return msr, dissatisfaction_counts, grid
    elif collect_frames:
        return grid, frames, moves
    else:
        return grid
````

````python
# Initialise simulation
init_rng = default_rng(seed)
init_grid = init_rng.choice([0,1,2], (grid_size, grid_size), p=state_probs)

# Run the simulation
sim_rng = default_rng(seed)
final_grid, frames, moves = run_one_simulation(init_grid, find_dissatisfied_indices, collect_data=False, collect_frames=True, sim_rng=sim_rng, max_iterations=200)
````

````python
def create_gif(frames, moves, filename, interval=400, figsize=(4, 4)):
    """
    Creates and displays a GIF animation of simulation frames for use in Jupyter slides.
    Requires a filename and overlays move indicators on each frame.

    Parameters:
    - frames: List of 2D arrays representing simulation steps.
    - moves: List of (from, to) tuples, with None for the first frame.
    - filename: Output filename for the GIF (e.g., 'schelling.gif').
    - interval: Delay between frames in milliseconds.
    - figsize: Size of the animation figure.
    """
    assert filename, "Filename must be provided."
    # Ensure it's saved in the same folder as the running notebook
    base_path = os.path.dirname(os.path.abspath("__file__"))  # fallback
    full_path = os.path.join(base_path, filename)


    cmap = ListedColormap(["white", "#FFD700", "#4682B4"])
    fig, ax = plt.subplots(figsize=figsize)

    def update(frame_index):
        ax.clear()
        ax.imshow(frames[frame_index], cmap=cmap, vmin=0, vmax=2)
        ax.set_xticks([]); ax.set_yticks([])

        move = moves[frame_index]
        if move:
            (i_from, j_from), (i_to, j_to) = move
            ax.plot(j_from, i_from, 'r*', markersize=12, label='from')
            ax.plot(j_to, i_to, 'k*', markersize=12, label='to')
            ax.legend(loc='upper left', frameon=True)

    ani = FuncAnimation(fig, update, frames=len(frames), interval=interval)
    ani.save(filename, writer='pillow')
    plt.close(fig)

    display(Image(filename=filename))
    return None  # Prevents object from being echoed in output
````

### Quantitative analysis

We want to measure the amount of segregation so we better actually define it.

##### *One* option for quantifying segregation
The mean similarity ratio:

$$
\mathcal{MSR}=\frac{1}{N}\sum_i^N \frac{n_s}{n},
$$

where $n_s$, $n_e$, $n$ denote the number of similar, empty and total neighbours in a Moore neighbourhood of the individual $i$. Hence, $n=9-1-n_e$

<font color='red'> This is what we will *choose*</font> (and what Schelling used). But there are many other choices.


In theory, the expected MSR for a perfectly mixed, infinite two-group grid is 0.5. In practice, smaller values can occur on finite grids (or with uneven populations).

We'd prefer our measure varied between 0 and 1, where 0 indicates a completely integrated and diverse population and 1 indicates complete segregation, because it's more directly comparable to other measures and easier to interpret (can normalise). 

````python
def get_mean_similarity(grid):
    """Average similarity ratio across all non-empty agents."""
    ratios, count = 0.0, 0

    rows, cols = grid.shape
    for idx in range(rows * cols):
        i, j = divmod(idx, cols)

        if grid[i, j] == 0:
            continue
        neighbourhood = grid[max(0, i-1):min(grid_size, i+2),
                             max(0, j-1):min(grid_size, j+2)]
        empty_n   = np.count_nonzero(neighbourhood == 0)
        tot_neigh = neighbourhood.size - empty_n - 1
        if tot_neigh > 0:
            similar = np.count_nonzero(neighbourhood == grid[i, j]) - 1
            ratios += similar / tot_neigh
            count  += 1
    return ratios / max(count, 1)
````

````python
# Let's package things up so it is easier to use...
def simulate_and_display(grid_size=10,
                         state_probs=[0.1, 0.45, 0.45],
                         threshold_percentage=50,
                         force_iterate=False,
                         simulate=True,
                         collect_data=False,
                         run_simulation=None,
                         max_iterations=200,
                         find_dissatisfied_indices=None,
                         get_mean_similarity=None,
                         init_rng=None,
                         sim_rng=None):
    """
    Run and display a Schelling simulation using interactive widgets.

    Parameters:
    - simulate: if False, will do nothing (can be used to skip in batch runs)
    - collect_data: if True, collects MSR and dissatisfaction history
    - run_simulation: function like run_one_simulation
    - find_dissatisfied_indices: function to find dissatisfied (must return (i, j) tuples)
    """

    assert run_simulation is not None, "Must provide a run_simulation function"
    assert find_dissatisfied_indices is not None, "Must provide find_dissatisfied_indices function"
    assert get_mean_similarity is not None, "Must provide get_mean_similarity function"

    if not simulate:
        print("Simulation skipped.")
        return

    if init_rng is None:
        init_rng = default_rng()
        print("WARNING: New RNG used for grid init")

    if sim_rng is None:
        sim_rng = default_rng()
        print("WARNING: New RNG used for simulation")
        
    # Initialise grid
    init_grid = init_rng.choice([0, 1, 2], (grid_size, grid_size), p=state_probs)

    # Run the simulation
    result = run_simulation(
        init_grid,
        find_dissatisfied_indices=find_dissatisfied_indices,
        get_mean_similarity=get_mean_similarity,
        threshold_percentage=threshold_percentage,
        force_iterate=force_iterate,
        collect_data=True,
        collect_frames=True, 
        sim_rng=sim_rng,
        max_iterations=max_iterations
    )

    # Unpack result based on what was collected
    if collect_data:
        msr, dissatisfaction_counts, final_grid, frames, moves = result
    else:
        final_grid, frames, moves = result

    # Prepare visualisation
    max_frame = len(frames) - 1
    cmap = ListedColormap(["white", "#FFD700", "#4682B4"])
    out = widgets.Output()

    def show_frame(frame_index):
        with out:
            clear_output(wait=True)
            fig, ax = plt.subplots(figsize=(4, 4))
            ax.imshow(frames[frame_index], cmap=cmap, vmin=0, vmax=2)
            ax.set_xticks([]); ax.set_yticks([])

            move = moves[frame_index]
            if move:
                (i_from, j_from), (i_to, j_to) = move
                ax.plot(j_from, i_from, 'r*', markersize=12, label='from')
                ax.plot(j_to, i_to, 'k*', markersize=12, label='to')
                ax.legend(loc='upper left', frameon=True)

            dissatisfied = find_dissatisfied_indices(frames[frame_index], threshold_percentage)
            for i, j in dissatisfied:
                ax.text(j, i, 'üòû', ha='center', va='center', fontsize=100 / grid_size, color='black')

            plt.tight_layout()
            plt.show()
            print(f"Frame {frame_index}: {len(dissatisfied)} dissatisfied agents")

    # Create slider and playback
    slider = widgets.IntSlider(value=0, min=0, max=max_frame, step=1, description='Time')
    play   = widgets.Play(interval=400, value=0, min=0, max=max_frame, step=1)  # Slower default
    widgets.jslink((play, 'value'), (slider, 'value'))
    slider.observe(lambda change: show_frame(change['new']) if 'new' in change else None, names='value')


    # Display
    # Create output containers
    show_frame(0)  # Draw first frame
    simulation_box = widgets.VBox([out, widgets.HBox([play, slider])])

    if collect_data:
        msr_out = widgets.Output()
        with msr_out:
            fig, ax = plt.subplots(figsize=(5, 4))
            plot_time_series(msr, ax=ax, ylabel="Mean similarity")
            # ax.set_title("Mean similarity over time")
            plt.tight_layout()
            plt.show()
        display(widgets.HBox([simulation_box, msr_out]))
    else:
        display(simulation_box)
````

````python
#time to steady state
def compute_time_to_steady_state(msr, epsilon=1e-3, window=5, max_iterations=200):
    diffs = np.abs(np.diff(msr))
    for i in range(len(diffs) - window):
        if np.all(diffs[i:i+window] < epsilon):
            return i + window
    return max_iterations

epsilon = 1e-3
window = 5

steady_times = {}
for threshold, msr_list in msr_results.items():
    times = [compute_time_to_steady_state(msr_list, epsilon=epsilon, window=window, max_iterations=max_iterations) 
             for msr in msr_list]
    steady_times[threshold] = {
        "mean": np.mean(times),
        "std": np.std(times)
    }

thresholds = sorted(steady_times.keys())
means = [steady_times[t]["mean"] for t in thresholds]
stds = [steady_times[t]["std"] for t in thresholds]

fig, ax = plt.subplots(figsize=(7, 5))
ax.errorbar(thresholds, means, yerr=stds, fmt='-o', capsize=4, color="darkblue")
ax.set_xlabel("Threshold (%)", fontsize=14)
ax.set_ylabel("Time to Steady State", fontsize=14)
# ax.set_title(f"Time to Steady State (Œµ = {epsilon}, window = {window})", fontsize=14)
ax.grid(True)
plt.tight_layout()
plt.show()
````

This is called an **ensemble** of simulations. It lets us report statistics, not specifics!

````python
def run_msr_ensemble(grid_size, state_probs, thresholds=None, seeds=range(5),
                     run_simulation_func=None, find_dissatisfied_indices=None,
                     get_mean_similarity=None):
    """
    Run an ensemble of Schelling model simulations across multiple thresholds.
    """
    if thresholds is None:
        thresholds = list(range(10, 100, 5))

    msr_ensemble_results = {}

    for threshold in tqdm(thresholds, desc="Thresholds"):
        all_runs = []
        for seed in tqdm(seeds, desc=f"Seed runs (T={threshold})", leave=False):
            init_rng = default_rng(seed)
            sim_rng = default_rng(seed)

            grid = init_rng.choice([0, 1, 2], (grid_size, grid_size), p=state_probs)

            msr, *_ = run_simulation_func(
                grid=grid,
                find_dissatisfied_indices=find_dissatisfied_indices,
                get_mean_similarity=get_mean_similarity,
                threshold_percentage=threshold,
                collect_data=True,
                collect_frames=False,
                force_iterate=True,
                sim_rng=sim_rng
            )
            all_runs.append(msr)
        
        msr_ensemble_results[threshold] = all_runs

    return msr_ensemble_results
````

````python
threshold_values = list(range(10, 100, 10))   # Example: 10% to 90% in steps of 10
seeds = range(5)                       # Number of ensemble runs per threshold


msr_ensemble_results = run_msr_ensemble(
    grid_size=10,
    state_probs=state_probs,
    thresholds=threshold_values,
    seeds=seeds,
    run_simulation_func=run_one_simulation,
    find_dissatisfied_indices=find_dissatisfied_indices,
    get_mean_similarity=get_mean_similarity
)
````

````python
thresholds = sorted(msr_ensemble_results)

# Prepare data
max_means = [np.mean([np.max(run) for run in msr_ensemble_results[t]]) for t in thresholds]
max_stds  = [np.std([np.max(run) for run in msr_ensemble_results[t]]) for t in thresholds]

time_means = [np.mean([np.argmax(run) for run in msr_ensemble_results[t]]) for t in thresholds]
time_stds  = [np.std([np.argmax(run) for run in msr_ensemble_results[t]]) for t in thresholds]

# Plot
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharex=True)


# Time to Max Segregation
axes[0].errorbar(thresholds, time_means, yerr=time_stds, fmt='s--', capsize=4, color='darkred')
# axes[0].set_title("Time to Max MSR vs Threshold", fontsize=14)
axes[0].set_xlabel("Threshold (%)", fontsize=14)
axes[0].set_ylabel("Time to Max MSR", fontsize=14)

# Max Segregation
axes[1].errorbar(thresholds, max_means, yerr=max_stds, fmt='o-', capsize=4, color='black')
# axes[1].set_title("Max MSR vs Threshold", fontsize=14)
axes[1].set_xlabel("Threshold (%)", fontsize=14)
axes[1].set_ylabel("Max Mean Similarity Ratio", fontsize=14)

# Finalise
plt.tight_layout()
plt.show()
````
