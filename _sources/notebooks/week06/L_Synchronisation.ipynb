
Synchronisation was first famously tackled by Dutch scientist Christiaan Huygens back in the 17th century as he attempted to solve the longitude problem for sailors. 

In an effort to improve the accuracy of timekeeping Huygens invented the pendulum clock and spoke of “the sympathy of two clocks” to describe their ability to synchonise. 

The synchronisation was due to weak coupling between the pendulums via the vibrations transmitted through the common wooden beam support (the fact that they synchronised ensured they were keeping the same time, which was not practically possible with two separate clocks).

# Oscillators
Dynamical systems can exhibit self-excited oscillations or periodic/repetitive behaviour.

Such systems are referred to as **oscillators**. e.g. a pendulum, or the internal clock of a firefly.

## A single oscillator
i.e. a point moving round a circle

Consider vector fields on the circle:

$$
\dot{\theta}=f(\theta),
$$

where $\theta$ is the angle, or phase, and determines where on the circle one is. 

By moving around the circle we can return to our starting point and so we have periodic solutions.


### Simple oscillators

$$
\dot{\theta}=\omega,
$$

where $\omega$ is a constant. 

Hence, $\theta=\omega t + \theta_0$. 

So simple oscillators have uniform, periodic motion around a circle at angular frequency $\omega$ and period $T=\frac{2\pi}{\omega}$ (we're ignoring amplitude).


<div style="border-left: 4px solid #1e70bf; padding: 0.75em; background-color: #eaf3fb; margin-bottom: 1em;">
  <strong>Check:</strong><br>
  Why exactly is a simple oscillator's motion uniform and periodic around a circle at angular frequency $\omega$ and period $T=\frac{2\pi}{\omega}$?
</div>

```{toggle}
Because $\dot\theta=\omega$ means the angle increases linearly in time:

$$
\theta(t)=\omega t+\theta_0.
$$

“Uniform” follows since $\mathrm{d}\theta/\mathrm{d}t$ is constant: equal angle per equal time.

The angle is mapped to the unit circle (or radius $A$) by:

$$
x(t)=A\cos\theta(t),\qquad y(t)=A\sin\theta(t).
$$

Then the state $(x,y)$ moves on the circle $x^2+y^2=A^2$. Its speed is constant:

$$
\dot{\mathbf r}(t)=\frac{d}{dt}
\begin{bmatrix}
A\cos\theta \\ A\sin\theta
\end{bmatrix}
=
\omega
\begin{bmatrix}
- A\sin\theta \\ A\cos\theta
\end{bmatrix},
\qquad
\|\dot{\mathbf r}(t)\|=A\omega=\text{const}.
$$

So the motion is uniform circular motion with no equilibrium unless trivially $\omega=0$.

And of course, it’s periodic because angles differing by $2\pi$ represent the same point on the circle. The smallest $T>0$ with $\theta(t+T)=\theta(t)+2\pi$ solves

$$
\omega T=2\pi \;\;\Rightarrow\;\; T=\frac{2\pi}{\omega},
$$

and any scalar “oscillator output” like $x(t)=A\cos(\omega t+\theta_0)$ or $y(t)=A\sin(\omega t+\theta_0)$ then satisfies $x(t+T)=x(t)$. Thus a simple (phase) oscillator corresponds to uniform rotation at angular frequency $\omega$ and period $2\pi/\omega$ (with amplitude $A$ fixed).
```

More complicated vector fields could be considered such as $f(\theta)=\sin(\theta)$, which has two fixed points $\theta^* = 0, \pi$.

## Two (uncoupled) simple oscillators
i.e. two points moving around a circle independently of each other:

![](images/Two_simple_oscillators.png)


### The dynamics of the phase difference

$$
\dot{\theta}_1 = \omega_1, \quad \dot{\theta}_2 = \omega_2.
$$

If the oscillators start aligned, then after some time their **phase difference** is

$$
\phi = \theta_1 - \theta_2,
$$

with dynamics

$$
\dot{\phi} = \dot{\theta}_1 - \dot{\theta}_2 = \omega_1 - \omega_2.
$$



This phase difference evolves at a constant rate. It behaves like another oscillator with period

$$
T = \frac{2\pi}{|\omega_1 - \omega_2|},
$$

describing the relative timing — whether the two oscillators remain locked, drift apart, or periodically realign. In particular, when $\phi$ increases by $2\pi$, oscillator 1 has lapped oscillator 2 once. Equivalently, this happens after

$$
T = \left(\frac{1}{T_1} - \frac{1}{T_2}\right)^{-1},
$$

where $T_i = \tfrac{2\pi}{\omega_i}$ is the natural period of oscillator $i$.


### Interference pattern when superimposing signals

If instead of analysing the phase difference you superimpose the signals, e.g.

$$
x(t) = \sin(\omega_1 t) + \sin(\omega_2 t),
$$

you obtain an **interference pattern**:

$$
x(t) = 2 \cos\!\left(\tfrac{\omega_1 - \omega_2}{2}\, t\right) \, \sin\!\left(\tfrac{\omega_1 + \omega_2}{2}\, t\right).
$$



This is the **beat phenomenon**. You hear a fast oscillation at the average frequency, with an amplitude envelope oscillating at the **beat frequency**

$$
f_{\text{beat}} = \frac{|\omega_1 - \omega_2|}{2\pi}.
$$

This envelope corresponds to the same relative frequency difference that drives the phase dynamics — but expressed in terms of *observable amplitude fluctuations* rather than angular drift.

</div>

```{toggle}
- **Phase difference** lives on the circle and tells you about relative timing.
- **Beats** live in the superimposed signal and tell you about audible (or visible) interference.

They are two different manifestations of the same underlying frequency mismatch.
```

<div style="border-left: 4px solid #1e70bf; padding: 0.75em; background-color: #eaf3fb; margin-bottom: 1em;">
  <strong>Check:</strong><br>

Derive the time $T$ after which Oscillator 1 laps Oscillator 2.
</div>

```{toggle}
The phase of each oscillator at time $t$ can be expressed as:
- Oscillator 1: $ \theta_1(t) = \frac{2\pi t}{T_1} $
- Oscillator 2: $ \theta_2(t) = \frac{2\pi t}{T_2} $

And hence phase difference at any time $t$ is given by:

$$
\Delta \theta(t) = \theta_1(t) - \theta_2(t) = \frac{2\pi t}{T_1} - \frac{2\pi t}{T_2} = 2\pi t \left( \frac{1}{T_1} - \frac{1}{T_2} \right)
$$

One oscillator laps the other when the phase difference reaches $2\pi$, meaning one oscillator has completed exactly one more cycle than the other. So, we set $\Delta \theta(t) = 2\pi$:

$$
2\pi t \left( \frac{1}{T_1} - \frac{1}{T_2} \right) = 2\pi
$$


Solving for $t$:

$$
t = \frac{1}{\frac{1}{T_1} - \frac{1}{T_2}} = \left( \frac{1}{T_1} - \frac{1}{T_2} \right)^{-1}
$$


```

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sampling_rate = 44100  # Sampling rate in Hz
duration = 5           # Duration in seconds
frequency_1 = 440      # Frequency of first oscillator (Hz)
frequency_2 = 442      # Frequency of second oscillator (Hz)

t = np.linspace(0, duration, int(sampling_rate*duration), endpoint=False)

# Oscillator signals
signal_1 = np.sin(2 * np.pi * frequency_1 * t)
signal_2 = np.sin(2 * np.pi * frequency_2 * t)

# Combined signal (beat phenomenon)
combined_signal = signal_1 + signal_2

# Plotting (zoom in on first 0.03 seconds for clarity)
xmax = 0.03
mask = t <= xmax

plt.figure(figsize=(6, 6))

# Overlay oscillator 1 & 2
plt.subplot(2, 1, 1)
plt.plot(t[mask], signal_1[mask], label='Oscillator 1 (440 Hz)', linestyle='-')
plt.plot(t[mask], signal_2[mask], label='Oscillator 2 (442 Hz)', linestyle='--', color='orange')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend(loc='upper right')

# Combined signal
plt.subplot(2, 1, 2)
plt.plot(t[mask], combined_signal[mask], label='Combined signal (beats)', color='green')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
````

This is *not* synchronisation.

A sinusoid sounds like a pure tone because it is just one frequency (no harmonics) and therefore has a flat envelope and constant amplitude. 

The oscillation is in air pressure, which our ears perceive as pitch (frequency), not as amplitude modulation (volume). Loudness is related to the amplitude squared (energy), which stays constant in time.

For the linearly combined signal:

\begin{align*}
x(t) &= \sin(2\pi f_1 t) + \sin(2\pi f_2 t)\\
    &= 2 \cos(2\pi \tfrac{\Delta f}{2} t) \cdot \sin(2\pi f_{\text{avg}} t),
\end{align*}

with $f_2 = f_1 + \Delta f$ and $f_{\text{avg}} = (f_1+f_2)/2$.

- The slow cosine envelope ($\Delta f / 2$) modulates the amplitude.
- The fast oscillation ($f_{\text{avg}}$) sets the perceived pitch.

To our ears, this sounds like a single note whose loudness “beats” up and down at frequency $|f_1-f_2|$.

# Synchronisation
Synchronisation occurs when **coupling** causes dynamical systems to lock their phases, maintaining a constant phase difference and typically a common frequency.

It is a key type of emergent collective phenomena that may be observed in complex systems.


Coupling can be the result of different types of interaction:
- **physical** e.g. mechanical systems or the metronomes that influence each other via the table they are on
- **informational** e.g. communication networks, where nodes transmit and receive data/signals
- **feedback** via a closed-loop interaction where the output of one system is fed back as an input to the other system e.g. thermostat-controlled heating system in a room where the output (room temperature) influences the input (heating system activation), which in turn affects the output 
- **functional** with behaviour of one system relying on the output or input of the other system. e.g. the heart and lungs in the circulatory system

The coupling can be uni-directional with one system driving the other or bi-directional with the systems mutually coupled

So what happens when two (or more) oscillators are coupled? Will we still observe beat phenomena? 

## Two coupled oscillators
i.e. that there is an interaction or influence between two points as they move around a circle, which affects their behaviour/dynamics.

![](images/Two_simple_oscillators_coupled.png)


Consider two oscillators, where the second oscillator drives the first. 

We want to model the following behaviour:
- If the first oscillator lags the second it will speed up
- If the first oscillator is ahead of the second it will slow down

i.e. they are coupled in a way that they would like to remain "close", despite having different intrinsic speeds.


One such coupling that achieves this is to modify the dynamics of the first oscillator to:

$$
\dot{\theta}_1 = \omega_1+K\sin(\theta_2-\theta_1)
$$

$K>0$ is a constant **coupling strength** that controls how strong the influence of the driving oscillator is.

The $\sin(\theta_2-\theta_1)$ measures closeness on the phase circle. The sign gives the direction of adjustment and its magnitude gives how strongly the phase is pulled.

## The dynamics of the phase difference

Again, we investigate: 

$$
\phi=\theta_2-\theta_1
$$

Then:

$$
\dot{\phi}=\dot{\theta}_2-\dot{\theta}_1=\omega_2-\omega_1-K\sin\phi.
$$

This is a non-uniform oscillator and so things get interesting.

### Nondimensionalisation

To simplify the analysis we can remove unnecessary units and make any equations more general.

For example, non-dimensionalised units of time let us express time in a system without using standard units like seconds, minutes, or hours. Instead, time is scaled by a characteristic time scale specific to the system under consideration.

We define dimensionless parameters:

$$
\tau=Kt
$$

$$
\mu=\frac{\omega_2-\omega_1}{K}
$$

<div style="border-left: 4px solid #1e70bf; padding: 0.75em; background-color: #eaf3fb; margin-bottom: 1em;">
  <strong>Theoretical Question:</strong><br>
  How should these dimensionless parameters be interpreted?
</div>

``` {toggle}
$\tau = Kt$: Time rescaled by the coupling strength $K$. This means the “natural unit of time” is set by how strongly the oscillators interact and one unit of $\tau$ corresponds to the timescale over which coupling effects matter. i.e. this is a dimensionless interaction time

$\mu = \tfrac{\omega_2 - \omega_1}{K}$: The frequency mismatch relative to coupling strength. It measures how large the detuning is compared to how strongly the oscillators pull on each other. i.e. this is the competition between detuning and coupling and tells you how big the natural frequency difference is relative to the coupling.

- If $|\mu| \ll 1$: coupling dominates the detuning, so the oscillators can synchronise (phase-lock).
- If $|\mu| \gg 1$: detuning dominates, so the oscillators drift apart and do not lock.

So $\mu$ is essentially the “competition parameter” that decides whether synchronisation is possible on the interaction timescale set by $\tau$.
```

We can now rewrite the dynamics of the phase difference in terms of our non-dimensionalised parameters: 

$$
\dot{\phi}=\dot{\theta}_2-\dot{\theta}_1=\omega_2-\omega_1-K\sin\phi
$$

Hence:

$$
\frac{\mathrm{d}\phi}{\mathrm{d}\tau}=\mu-\sin\phi.
$$

So what do we do with this...?

### Classifying the behaviour (if you've done MATH3021)

If you've done MATH3021 you wouldn't dream of trying to analytically solve this equation:

$$
\frac{\mathrm{d}\phi}{\mathrm{d}\tau}=\mu-\sin\phi.
$$

Instead, you'd sketch the phase portrait of $\phi$ vs $\dot{\phi}=\mu-\sin\phi$ and analyse its fixed points and their stability

The fixed points exist when $\mu=\sin\phi$:
   - If $ |\mu| \leq 1 $, there are two fixed points given by $ \phi = \arcsin(\mu) $ and $\phi = \pi - \arcsin(\mu) $.
   - If $|\mu| > 1$, there are no real fixed points because $\sin\phi$ can only take values in the interval $[-1, 1]$.

To determine the stability of the fixed points we linearise the equation around each fixed point by computing the derivative of the right-hand side with respect to $\phi$:

$$
\frac{\mathrm{d}}{\mathrm{d}\phi}\left(\mu - \sin\phi\right) = -\cos\phi.
$$

At the fixed points $\phi^*$ where $\sin\phi^* = \mu$, the stability is determined by the sign of $-\cos\phi^*$:

- **Stable Fixed Point**: If $-\cos\phi^* < 0$ (i.e., $\cos\phi^* > 0$), the fixed point is **stable**. i.e. perturbations around this point will decay over time, leading the system back to the fixed point.
- **Unstable Fixed Point**: If $-\cos\phi^* > 0$ (i.e., $\cos\phi^* < 0$), the fixed point is **unstable**. i.e. perturbations will grow over time, causing the system to move away from the fixed point.


Interpreting this in the context of synchronisation:
- **Complete synchronisation**: oscillators oscillate with the same frequency and have identical phases $\phi=n\pi$, where $n$ is even
- **Phase-locked synchronisation**: oscillators maintain a constant phase difference relative to each other but may have different frequencies
    - this will occur for all other values of $\phi$
- **Phase drift**: a specific form of partial synchronisation where the phases of the oscillators exhibit a constant phase difference that changes at a steady rate
    - will occur if $|\mu|>1$ as there are no fixed points
- **Unsynchronised**: no coherent phase relationship, leading to desynchronised behavior where oscillators operate nearly independently of each other
    - will not occur in this case

# Networks of coupled oscillators

Why stop at two? We could couple an entire network of oscillators. 

Generically, a coupled oscillator model can be modelled as:

$$
\frac{\mathrm{d} c_i}{\mathrm{d} t} = l_i(c_i) + k_i \sum_{j\in N_i} f(c_j - c_i).
$$

Here $l_i$ are the local/internal dynamics of oscillator $i$, $f$ is the coupling function.

e.g. for diffusion $f$ is typically a simple difference in concentrations $c$ and $k_i$ are called diffusion constants. 

## Kuramoto model

We will now return to the Kuramoto model and study the emergent synchronisation in more quantitative detail.

This is a model of a system of oscillators that can synchronise despite differences in the oscillators' individual natural frequencies (their local dynamcis $l_i$).

Specifically, the Kuramoto model is defined for $N$ oscillators with *equally weighted all-to-all coupling*:

$$ 
\dot{\theta}_i = \omega_i + \frac{K}{|\mathcal{N}_i|} \sum_{j \in \mathcal{N}_i} W_{ij}\sin(\theta_j - \theta_i).
$$



- $\omega_i$ is the intrinsic/natural frequency of the individual oscillator drawn from a unimodal distribution
- $\sin(x)$ is the coupling function
- we average over the number of neighbours $i$ has, $|\mathcal{N}_i|$ (here a fully connected network topology is used so $|\mathcal{N}_i|=N$)
- $K$ is simply a global coupling strength
- $W_{ij}=1/N$ are coupling weights

The model makes several assumptions e.g. weak coupling, identical (or nearly identical) oscillators, interactions that depend sinusoidally on phase differences between each pair of objects, etc.

## Quantitative analysis
To quantify and summarise the amount of synchronisation we compute a complex order parameter:

$$
r(t)e^{\mathrm{i}\psi(t)} = \frac{1}{N} \sum_{j=1} e^{\mathrm{i} \theta_j(t)}.
$$

It is effectively a way to find an average angle avoiding the awkwardness of doing things mod $2\pi$ by considering vectors in the complex plane.

$r=|re^{\mathrm{i}\psi}|$ is a measure of the phase coherence. 
- $r\approx 1 \implies $ the oscillators are 'clumped' together (synched) 
- $r \approx 0 \implies $ the oscillators are scattered around the circle (independent)
- $r_{\infty}$ is the steady-state coherence $r$ (at $t=\infty$)
- $\psi$ is like an average phase.


We've seen this before when we looked at quantifying the global behaviour of a flock. It is an order parameter here too and is measure of the average behaviour of the system that quantifies the extent to which the oscillators are synchronised.

### Mean-field analysis 
In the Kuramoto model, the interaction network is a complete graph, which allows a nice mean-field analysis (via a uniform average and  uniform neighbour set/weights).



Multiplying the full Kuramoto order parameter ($r e^{\mathrm{i}\psi}$) by $e^{-\mathrm{i}\theta_i}$, expanding via Euler's formula and taking the imaginary part gives:

$$
re^{\mathrm{i}(\psi−\theta_i)} = \frac{1}{N}\sum_j^N e^{\mathrm{i} (\theta_j - \theta_i)} 
$$


$$
\text{Im}(re^{\mathrm{i}(\psi−\theta_i)})=r\sin(\psi-\theta_i)=\frac{1}{N}\sum_{j=1}^N\sin(\theta_j-\theta_i)
$$



Hence the $N$-body system on an all-to-all network can be written *exactly* as:

\begin{align*}
\dot{\theta}_i &= \omega_i + \frac{K}{N} \sum_{j \in \mathcal{N}_i} \sin(\theta_j - \theta_i) \\
&=  \omega_i + K\,r\,\sin(\psi-\theta_i),
\qquad i=1,\dots,N.
\end{align*}

In this form each oscillator can be seen to be essentially uncoupled from the others with dynamics controlled by only the two mean-field quantities: the coherence $r\in[0,1]$ and mean phase $\psi$. i.e. The phase $\theta_i$ is pulled towards the mean phase $\psi$ with effective strength proportional to the coherence $r$.


If you've done MATH3021 you could now analyse this simple ODE:
- The fixed points satisfy $\omega_i=Kr\sin(\theta_i)$ and exist if $|\omega_i|\leq Kr$ (and one is stable/phase-locked, from the stability analysis)
- if $|\omega_i|\leq Kr$ then we have phase-drift.

## Numerical analysis
$K$ is the parameter of interest.

Integrating the model numerically we can observe how $r$ evolves in time, for a particular value of $K$:

![](images/Coherence_evolution.png)

Image from: Fig. 2 of [Strogatz, 2000](https://doi.org/10.1016/S0167-2789(00)00094-4)



- $K<K_c$, the oscillators act as if they were uncoupled with $r_{\infty} \approx 0$: the phases become uniformly distributed around the circle, starting from any initial condition
- $K\approx K_c$, the incoherent state becomes unstable and $r(t)$ grows exponentially as a small cluster of oscillators become synchronised generating a collective oscillation. 
    - population splits into two groups: oscillators near center of frequency distribution lock at mean frequency and co-rotate with average phase, while those in tails run near their natural frequencies and drift relative to synchronised cluster. Often called **partially synchronised**. 
- $K>K_c$ more and more oscillators are recruited into the synchronised cluster, and $r_{\infty}$ grows and eventually stabilises at approximately 1 for full synchronisation.

Alternatively, we might be more interested in viewing the final coherence, $r_{\infty}$.

The long-time order parameter demonstrates a second-order (continuous) phase transition in the behaviour for varied $K$. Proper ensemble runs demonstrate that $r_{\infty}$ is independent of the initial condition, i.e. there is a globally attracting state for each value of $K$.

![](images/Param_sweep.png)

Returning to the [Kuramoto Complexity Explorable](https://www.complexity-explorables.org/explorables/ride-my-kuramotocycle/), we can now describe the observed modes more precisely:

**Unsynchronised (incoherent)**
- Phases spread around the circle with no persistent phase relation; $r\approx 0$
- Typical with weak coupling or a broad spread of natural frequencies

**Drifting (quasi-periodic)**
- Not locked; phase differences wander in time and never settle. Order parameter small but nonzero and fluctuating
- Occurs when coupling is too weak to overcome detuning; partial, transient clusters appear and dissolve

**Phase-locked (frequency synchrony)**
- Oscillators share a common frequency with **constant** phase offsets (not necessarily zero); $r$ is high (less than 1 if offsets vary)
- Achieved when coupling overcomes frequency heterogeneity (or trivially if intrinsic frequencies are nearly identical)

**Complete synchronisation (identical phases)**
- All phases coincide (offsets $=0$ mod $2\pi$); $r=1$ 
- Requires sufficiently strong coupling and/or virtually identical intrinsic dynamics
