**Kepler's observation:** The orbit of every planet is an ellipse with the Sun at one of the two foci. (This, and his other two laws, were found by analysing astronomical data.)

**The model:** Isaac Newton proved Kepler was right by following a sort of 'recipe'... 
   1. he appealed to a general principle - the *law* of gravitation, which can be 
   2. written as a *differential equation* and 
   3. *analytically solved* this to give an elliptic solution
   4. *QED*. 

This mathematical derivation gives rigor and a sense of absolute and unwavering proof of the observed phenomena, so people tended to like this

Is it perfect?

No - ignores some stuff: forces of say other planets or relativistic effects, which we justify because they are small relative to those we deem important, size of planet, modelled as point masses as actual size is negligible relative to inter-planet distance. But these assumptions/approximations seem reasonable and justified.

# Complex Systems as a field

Complex Systems is essentially **a diverse and adaptable toolkit** for dealing with complex systems 

Note that I will typically capitalise the discipline and use lower case for a system that is complex.

'Classical' modelling describes the extreme scenarios well...

- **Simple systems:** comprised of a small number of interacting components that can be easily described using mathematical models.
    - e.g. motion of planets
    - described well by classical mechanics

- **Complicated (disorganised) systems:** comprised of a large number of mostly non-interacting components and are amenable to aggregation and linear modelling. 
    - e.g. The billions of atoms in a gas, but their interactions are trivial 
    - statistical mechanics describes their macro-scale behaviour well


These classical systems are normally closed, reversible, symmetry-dominated and predictable and we have very good theory to describe these systems and their behaviour

On the other hand, classical modelling can struggle when attempting to disentangle **complex systems** sitting between these extremes. 

... and that's exactly where most of the interesting stuff lies!

**Organised complex systems:** are composed of large numbers of parts (as in the disorganised systems) with non-trivial (typically nonlinear) interactions spanning multiple scales

They are characterised by properties such as...
- open (positioned within some gradient, e.g. chemical)
- self-organising (with rules that amplify fluctuations into coordinate space-time patterns)
- dissipative (dynamics that are irreversible and break time-reversal symmetry)
- uncertain (open-ended space of possible states not enumerable in initial conditions)
- adaptive (capture of information by a system from the environment that promostes persistence and increases multiplicity)

At one extreme there are fundamental regularities described using minimal assumptions, simple rules and few initial conditions: symmetry and law 

At the other there are emergent regularities constructed from contingent histories described with coarse-grained rules and nested boundary conditions: information and adaptation

`There are laws of physics and there are rules of life -- and rules are meant to be broken' - Volume 1 page 6

The complex domain encompasses the world of self-organising and evolving agents at all scales: from organisms to societies

#### Key features of complex systems

- **Interactions:** across spatial and/or temporal scales. 
- **A large number of components** (but not too large)
- **Nonlinearity:** Relationships between inputs and outputs, or between components within the system, are complex and often not linear.
    <!-- **Schelling:** interactions and feedback loops. Small changes in neighbourhood or preference can have large effect. -->
- **Heterogeneity:** Diversity across the components in the system such as different attributes, behaviours, or states.
    <!-- **Schelling:** small amount of heterogeneity with different 'agent' types (blue and yellow) and varied degrees of satisfaction. More could be included. -->
- **Emergence:** A theory of the system at some scale or level of organisation is qualitatively different from the lower level scale. i.e *the whole is more than the sum of its parts* - Aristole. Society is more than a collection of individuals, consciousness is more than a collection of neurons, traffic is more than a collection of cars. This is a behavioural phenomena at the system level
    <!-- **Schelling:** contrary to intuition, a high level of homophily (*homós:* 'same, common', *philía:* 'friendship, love') can produce a mixed state of the society because agents keep moving without reaching a stationary state, and it doesn't take much homophily to cause segregation to occur -->
- **Multiscale dynamics:** Dynamics occur across multiple scales. The system often lacks of a characteristic scale. How do we go from the dynamics of neurons to brain function and behaviour?
    <!-- **Schelling:** individual agent-level decisions contribute to the emergence of segregation at various scales. —>
- **Far from equilibrium:** complex systems usually exist in an open environment from which individual components can both consume and dissipate energy, allowing for time reversal symmetry to break. A complex system such as this is often referred to as 'active matter', if the individuals in the matter are moving somehow then they may be referred to as 'self-propelled particles'. Individuals may also have an ability to express a strategy, adapting, coordinating, optimising as they make decisions in an evolving environment.

- **Strong transitions:** Abrupt shifts can occur as systems pass critical points (called bifurcation or tipping points). This may continue to produce a cascade, where one failure (or success) can lead to cascading failures (or successes) that may have catastrophic (or beneficial) consequences.

### Changing paradigm
In both science and philosophy, a paradigm refers to a unique framework consisting of concepts, thought patterns, theories, research methods, postulates, and criteria that define what is considered valid and valuable within a particular field of study

Complex Systems is a relatively 'new' field, being coined circa 1980. It is new in that the tools, the thinking *and the criteria for judging a model* are new. It is at the centre of a gradual shift in what we consider to be (good) science... and good modelling.

It is one of the most radical new scientific paradigms of the twentieth century, forging four areas of research into a new way of seeing the world: entropy, evolution, dynamics and computation. Complex systems -- the theory of far-from-equilibrium, purposeful machines

The sorts of questions we ask of these systems: optimisation, robustness, resilience, stability, energy dissipation, regulatory mechanics, origination, informational constraints, computational costs and limits
