````python
# Test cell: Basic setup check

import numpy as np
import matplotlib.pyplot as plt

print("✅ All core packages imported successfully!")

# Simple plot: logistic map
r = 3.7
x = 0.5
xs = []
for _ in range(100):
    x = r * x * (1 - x)
    xs.append(x)

plt.plot(xs)
plt.title("Logistic Map Iteration")
plt.xlabel("Time step")
plt.ylabel("x")
plt.grid(True)
plt.show()
````

## Pseudo-code 
Start with pseudo-code!

Pseudo-code is language-agnostic so you to outline the logic and concentrate on the problem-solving process itself, rather than getting bogged down in sytax specifics.

You can write pseudo-code in many ways, as long as it is unambiguous. 

For example:
```
// Input: Two numbers, num1 and num2

// Step 1: Read the first number (num1)
Read num1

// Step 2: Read the second number (num2)
Read num2

// Step 3: Calculate the sum of num1 and num2
sum = num1 + num2

// Step 4: Display the result
Display sum
```

An alternative:

```
          +-----------------+
          | Start           |
          +-----------------+
                   |
                   |
          +--------v--------+
          | Read num1       |
          +--------+--------+
                   |
                   |
          +--------v--------+
          | Read num2       |
          +--------+--------+
                   |
                   |
          +--------v--------+
          |    sum = num1   |
          |    + num2       |
          +--------+--------+
                   |
                   |
          +--------v--------+
          |  Display sum    |
          +--------+--------+
                   |
                   |
          +--------v--------+
          |    Stop         |
          +-----------------+ 
```

Another alternative:
```
1. Start the program.
2. Prompt the user to enter the first number (num1).
3. Read the value of num1 from the user.
4. Prompt the user to enter the second number (num2).
5. Read the value of num2 from the user.
6. Calculate the sum of num1 and num2, and store it in a variable called sum.
7. Display the value of sum to the user.
8. End the program.
```
